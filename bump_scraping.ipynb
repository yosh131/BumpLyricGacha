```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time 
import re
from datetime import datetime
```

```python

list_df = pd.DataFrame(columns=["title", "lyric"])
url="https://www.uta-net.com/artist/126/"
base_url = "https://uta-net.com"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
all_a_tags = soup.find_all('a')
all_links = [a['href'] for a in all_a_tags if 'href' in a.attrs]
song_links = [l for l in all_links if l.startswith("/song/")]

# song_links.sort()
# print(song_links)


# song_links = song_links[127:128]


for link in song_links:
    song_link = base_url+link
    response_s = requests.get(song_link)
    soup_s = BeautifulSoup(response_s.text, 'lxml')
    
    song_name = soup_s.find('h2').text
    print(song_name)

    song_lyrics = soup_s.find('div', itemprop='lyrics')
    song_lyric = song_lyrics.get_text(separator='\u3000')
    song_lyric = song_lyric.replace('\n','　')
    song_lyric = song_lyric.replace('この歌詞をマイ歌ネットに登録 >', '')
    song_lyric = song_lyric.replace('このアーティストをマイ歌ネットに登録 >', '')

    print(song_lyric)
    new_row_data = {"title":song_name, "lyric":song_lyric}
    list_df.loc[len(list_df)] = new_row_data
    
    if song_name=="ロストマン":
        break
    time.sleep(1)
    



list_df.to_csv("bump_all_lyrics.csv", index=False, encoding="utf-8")
```

```python
import numpy as np
import pandas as pd
import re
from datetime import datetime

NUM_PHRASE=10

df = pd.read_csv("./bump_all_lyrics_fixed.csv")
MAX_NUM = len(df)

def get_lyric(df, seed=None):
    np.random.seed(seed=seed)

    idx = np.random.randint(0, MAX_NUM)
    lyric_all = df.iloc[idx,1]
    lyric_list = re.split(r'[ 　]+', lyric_all)
    idx_lyric = np.random.randint(0, len(lyric_list)-NUM_PHRASE+1)
    ret_lyric = lyric_list[idx_lyric]
    return ret_lyric


L = []
for _ in range(NUM_PHRASE):
    lyc = get_lyric(df)
    L.append(lyc)
print(*L)
```
