```python
import numpy as np
import pandas as pd
import re
from datetime import datetime

NUM_PHRASE=10

df = pd.read_csv("./bump_all_lyrics_fixed.csv")
MAX_NUM = len(df)

def get_lyric(df, seed=None):
    np.random.seed(seed=seed)

    idx = np.random.randint(0, MAX_NUM)
    lyric_all = df.iloc[idx,1]
    lyric_list = re.split(r'[ 　]+', lyric_all)
    idx_lyric = np.random.randint(0, len(lyric_list)-NUM_PHRASE+1)
    ret_lyric = lyric_list[idx_lyric]
    title = df.iloc[idx,0]
    return ret_lyric, title


L = []
T = []
for _ in range(NUM_PHRASE):
    lyc, title = get_lyric(df)
    L.append(lyc)
    T.append(title)
print(*L)
print(*T)
```
